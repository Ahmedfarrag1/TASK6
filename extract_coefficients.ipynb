```python
from scipy import signal
import numpy as np
```

```python
#continous transfer function
#using the given attributes: the cutoff freq = 21.4

cutoff_frequency = 21.4
w0 = 2*np.pi*cutoff_frequency   #the used formula: 2*PI*cutoff_frequency
num = w0
den = [1, w0]
lowPass = signal.TransferFunction(num, den)
print(lowPass)
```

```python
#discrete transfer functon
dt = 1.0/1000.0
discretelowPass = lowPass.to_discrete(dt, method = "gbt", alpha = 0.5)
print(discretelowPass)
```

```python
#difference equation coefficients
b = discretelowPass.num
a = -discretelowPass.den
print("filter coefficients bi: " + str(b))
print("filter coefficients ai: " + str(a[1:]))

```
